```python
import numpy as np
import pandas as pd
s1_walking = pd.read_csv('drive/MyDrive/IE5374 Workspace/Colab Notebooks/Project_2/Data/s1_walking_csv/acc_walking_head.csv')
s1_running = pd.read_csv('drive/MyDrive/IE5374 Workspace/Colab Notebooks/Project_2/Data/s1_walking_csv/acc_walking_head.csv')
s1_climbingup = pd.read_csv('drive/MyDrive/IE5374 Workspace/Colab Notebooks/Project_2/Data/s1_walking_csv/acc_walking_head.csv')
s1_climbingdown = pd.read_csv('drive/MyDrive/IE5374 Workspace/Colab Notebooks/Project_2/Data/s1_walking_csv/acc_walking_head.csv')
```

```python
# Find the resultant vector of x,y,z
#walking
res_vec_sq = (s1_walking['attr_x']**2)+(s1_walking['attr_y']**2)+(s1_walking['attr_z']**2)
res_vec = res_vec_sq**(1/2)
s1_walking['Res_vec']=res_vec
#Running
res_vec_sq = (s1_running['attr_x']**2)+(s1_running['attr_y']**2)+(s1_running['attr_z']**2)
res_vec = res_vec_sq**(1/2)
s1_running['Res_vec']=res_vec
#Climbing up
res_vec_sq = (s1_climbingup['attr_x']**2)+(s1_climbingup['attr_y']**2)+(s1_climbingup['attr_z']**2)
res_vec = res_vec_sq**(1/2)
s1_climbingup['Res_vec']=res_vec
#Climbing down
res_vec_sq = (s1_climbingdown['attr_x']**2)+(s1_climbingdown['attr_y']**2)+(s1_climbingdown['attr_z']**2)
res_vec = res_vec_sq**(1/2)
s1_climbingdown['Res_vec']=res_vec
```

```python
from ts2vg import HorizontalVG
g = HorizontalVG()
g.build(s1_walking['Res_vec'][1000:2048])
ig_g = g.as_igraph()
```
